```python
import pandas as pd
```

```python
epz = pd.read_csv('/content/empz_01.csv')
```

```python
epz.head(7)
```

```python
from sklearn.preprocessing import LabelEncoder
l_e = LabelEncoder()
```

```python
emp = epz.dropna()
```

```python
epz.info()
```

```python
dep_enc = l_e.fit_transform(emp.Department)
print(dep_enc)
```

```python
biz_tvl = l_e.fit_transform(emp.BusinessTravel)
print(biz_tvl)
```

```python
atr = l_e.fit_transform(emp.Attrition)
print(atr)
```

```python
ftr = list(zip(dep_enc,biz_tvl,atr))
```
