```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
```

```python
epz = pd.read_csv('/content/empz_01.csv')
```

```python
epz.columns
```

```python
x = epz[['StandardHours','DistanceFromHome']]
```

```python
x.head()
```

```python
from sklearn.linear_model import LinearRegression
mod = LinearRegression()
```

```python
y = epz[['YearsAtCompany']]
```

```python
mod.fit(x,y)
```

```python
mod.coef_
```

```python
mod.intercept_
```

```python
yr = mod.predict([[8,12]])
```

```python
print(yr)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(y_test)
```

```python
mod.fit(X_train,y_train)
```

```python
p_y = mod.predict(X_test)
```

```python
print(p_y)
```

```python
from sklearn import metrics
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test,p_y)))
```
